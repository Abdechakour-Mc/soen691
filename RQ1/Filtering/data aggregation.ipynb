# Pre Era

```python
import pandas as pd
```

```python
pre_info = pd.read_csv('../Data Collection/spider_output/javascript_projects_jan2018_active2021.csv')
pre_info.drop_duplicates(subset=['name'], inplace=True)
pre_info
```

```python
# Load ow results
ow_pre_df1 = pd.read_csv('../Scanning/owasp_reports/pre2019/dependency_check_summary.csv')
ow_pre_df2 = pd.read_csv('../Scanning/owasp_reports/pre2020/dependency_check_summary.csv')
ow_pre_df3 = pd.read_csv('../Scanning/owasp_reports/pre2021/dependency_check_summary.csv')
len(set(ow_pre_df1.projectName)), len(set(ow_pre_df2.projectName)), len(set(ow_pre_df3.projectName)),
```

```python
# Load sq results
sq_pre_df1 = pd.read_csv('../Scanning/sonar_reports/pre2019/summary.csv')
sq_pre_df2 = pd.read_csv('../Scanning/sonar_reports/pre2020/summary.csv')
sq_pre_df3 = pd.read_csv('../Scanning/sonar_reports/pre2021/summary.csv')
len(set(sq_pre_df1.project)), len(set(sq_pre_df2.project)), len(set(sq_pre_df3.project)), 
```

```python
import random

# Extract common projects
comm_projects = list(set(set(sq_pre_df1.project).intersection(set(sq_pre_df2.project))).intersection(set(sq_pre_df3.project)).intersection(set(ow_pre_df1.projectName)))
print(len(comm_projects))

# Assuming you have a list called 'my_list'
comm_projects = random.sample(comm_projects, 200)
print(len(comm_projects))
```

```python
len(set(pre_info.name).intersection(set(comm_projects)))
```

```python

ow_pre_df1 = ow_pre_df1[ow_pre_df1['projectName'].isin(comm_projects)]
ow_pre_df2 = ow_pre_df2[ow_pre_df2['projectName'].isin(comm_projects)]
ow_pre_df3 = ow_pre_df3[ow_pre_df3['projectName'].isin(comm_projects)]

sq_pre_df1 = sq_pre_df1[sq_pre_df1['project'].isin(comm_projects)]
sq_pre_df2 = sq_pre_df2[sq_pre_df2['project'].isin(comm_projects)]
sq_pre_df3 = sq_pre_df3[sq_pre_df3['project'].isin(comm_projects)]
```

```python
# Function to convert "XhYmin" or "Ymin" or "Xh" to total minutes
def convert_to_minutes(time_str):
    if 'h' in time_str and 'min' in time_str:
        # Case: XhYmin (e.g., 8h20min)
        hours, minutes = time_str.split('h')
        minutes = minutes.replace('min', '')
        return int(hours) * 60 + int(minutes)
    elif 'h' in time_str:
        # Case: Xh (e.g., 9h)
        hours = time_str.replace('h', '')
        return int(hours) * 60
    elif 'min' in time_str:
        # Case: Ymin (e.g., 25min)
        minutes = time_str.replace('min', '')
        return int(minutes)
    else:
        # Case: Handle unexpected formats (e.g., return 0 or raise an error)
        return 0  # Or raise ValueError("Invalid time format")
    
# Merge the filtered DataFrames on `project` and `projectName`
merged_pre_df1 = pd.merge(
    ow_pre_df1,
    sq_pre_df1,
    left_on='projectName',
    right_on='project',
    how='inner'  # Use 'inner' to keep only matching rows
).merge(
    pre_info,
    left_on='projectName',  # Use the appropriate column from the merged result
    right_on='name',
    how='inner'
)

merged_pre_df2 = pd.merge(
    ow_pre_df2,
    sq_pre_df2,
    left_on='projectName',
    right_on='project',
    how='inner'  # Use 'inner' to keep only matching rows
).merge(
    pre_info,
    left_on='projectName',  # Use the appropriate column from the merged result
    right_on='name',
    how='inner'
)

merged_pre_df3 = pd.merge(
    ow_pre_df3,
    sq_pre_df3,
    left_on='projectName',
    right_on='project',
    how='inner'  # Use 'inner' to keep only matching rows
).merge(
    pre_info,
    left_on='projectName',  # Use the appropriate column from the merged result
    right_on='name',
    how='inner'
)


merged_pre_df1['totalTimeNeededToRemoveDebts_mins'] = merged_pre_df1['totalTimeNeededToRemoveDebts'].apply(convert_to_minutes)
merged_pre_df2['totalTimeNeededToRemoveDebts_mins'] = merged_pre_df2['totalTimeNeededToRemoveDebts'].apply(convert_to_minutes)
merged_pre_df3['totalTimeNeededToRemoveDebts_mins'] = merged_pre_df3['totalTimeNeededToRemoveDebts'].apply(convert_to_minutes)


# Drop unuseful cols
merged_pre_df1 = merged_pre_df1.drop(columns=['project', 'numberOfVulnerabilities', 'totalTimeNeededToRemoveDebts', 'numberOfIssuesRelatedToMaintainability', 'numberOfIssuesRelatedToReliability', 'numberOfIssuesRelatedToSecurity', 'numberOfIssuesWithLowSeverity', 'numberOfIssuesWithMediumSeverity', 'numberOfIssuesWithHighSeverity', 'name', 'id'])
merged_pre_df2 = merged_pre_df2.drop(columns=['project', 'numberOfVulnerabilities', 'totalTimeNeededToRemoveDebts', 'numberOfIssuesRelatedToMaintainability', 'numberOfIssuesRelatedToReliability', 'numberOfIssuesRelatedToSecurity', 'numberOfIssuesWithLowSeverity', 'numberOfIssuesWithMediumSeverity', 'numberOfIssuesWithHighSeverity', 'name', 'id'])
merged_pre_df3 = merged_pre_df3.drop(columns=['project', 'numberOfVulnerabilities', 'totalTimeNeededToRemoveDebts', 'numberOfIssuesRelatedToMaintainability', 'numberOfIssuesRelatedToReliability', 'numberOfIssuesRelatedToSecurity', 'numberOfIssuesWithLowSeverity', 'numberOfIssuesWithMediumSeverity', 'numberOfIssuesWithHighSeverity', 'name', 'id'])
```

```python
merged_pre_df3.describe()
```

```python
# Exporting
merged_pre_df1.to_csv('./filtered data/pre_2019.csv', index=False)
merged_pre_df2.to_csv('./filtered data/pre_2020.csv', index=False)
merged_pre_df3.to_csv('./filtered data/pre_2021.csv', index=False)
```
