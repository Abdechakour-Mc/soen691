```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib import rcParams

# Set random seed for reproducibility
np.random.seed(42)

# Set style for better looking plots
sns.set_style("whitegrid")
rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

def load_and_sample_data(pre_file, post_file, sample_size=160):
    """Load both CSVs and sample the same number of projects from each"""
    pre_df = pd.read_csv(pre_file)
    post_df = pd.read_csv(post_file)
    
    # Sample the data
    pre_sample = pre_df.sample(min(sample_size, len(pre_df)), random_state=42)
    post_sample = post_df.sample(min(sample_size, len(post_df)), random_state=42)
    
    return pre_sample, post_sample

def get_sorted_categories(pre_data, post_data):
    """Get alphabetically sorted list of category names"""
    categories = [col.replace(' %', '') for col in pre_data.columns if '%' in col]
    return sorted(categories)

def plot_category_comparison(pre_data, post_data):
    """Compare category distributions between pre and post with consistent ordering"""
    # Get sorted categories
    categories = get_sorted_categories(pre_data, post_data)
    category_cols = [f"{cat} %" for cat in categories]
    
    # Calculate means and add "No Issue" category
    pre_means = pre_data[category_cols].mean()
    post_means = post_data[category_cols].mean()
    
    # Calculate remaining percentage for "No Issue"
    pre_no_issue = max(0, 100 - pre_means.sum())  # Ensure non-negative
    post_no_issue = max(0, 100 - post_means.sum())
    
    # Create comparison dataframe with "No Issue"
    comparison_df = pd.DataFrame({
        'Category': categories + ['No Issue'],
        'Pre': list(pre_means.values) + [pre_no_issue],
        'Post': list(post_means.values) + [post_no_issue]
    })
    
    # Melt for seaborn
    melted_df = pd.melt(comparison_df, id_vars='Category', var_name='Period', value_name='Percentage')
    
    # Professional color palette (Tol muted color scheme)
    palette = {
        'Pre': '#6495ED',  
        'Post': '#F28500'  # Golden
    }
    
    # Plot with consistent category order
    plt.figure(figsize=(14, 8))
    ax = sns.barplot(
        x='Category', 
        y='Percentage', 
        hue='Period', 
        data=melted_df,
        order=categories + ['No Issue'],
        palette=palette,
        edgecolor='white',
        linewidth=1
    )
    
    # Styling
    plt.title('Average Issue Category Distribution Comparison (Pre vs Post)', 
             pad=20, fontsize=14, fontweight='bold')
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.yticks(fontsize=10)
    plt.ylabel('Average Percentage (%)', fontsize=12)
    plt.xlabel('Issue Category', fontsize=12)
    
    # Custom legend
    plt.legend(title='Analysis Period', frameon=True, 
              facecolor='white', framealpha=1,
              bbox_to_anchor=(1.02, 1), loc='upper left')
    
    # Add value labels with improved formatting
    for p in ax.patches:
        height = p.get_height()
        if height > 0.5:  # Only label if > 0.5%
            ax.annotate(f"{height:.1f}%", 
                       (p.get_x() + p.get_width() / 2., height), 
                       ha='center', va='center', 
                       xytext=(0, 5), 
                       textcoords='offset points',
                       fontsize=9,
                       bbox=dict(boxstyle='round,pad=0.2', 
                                fc='white', 
                                ec='none', 
                                alpha=0.7))
    
    # Remove spines for cleaner look
    sns.despine(left=True, bottom=True)
    
    plt.tight_layout()
    plt.savefig('charts/category_comparison.png', dpi=300, bbox_inches='tight')
    plt.show()

def plot_category_pie_charts(pre_data, post_data):
    """Create pie charts for category distributions with consistent ordering"""
    # Get sorted categories
    categories = get_sorted_categories(pre_data, post_data)
    category_cols = [f"{cat} %" for cat in categories]
    
    pre_means = pre_data[category_cols].mean()
    post_means = post_data[category_cols].mean()
    
    # Calculate remaining percentage for "No Issue"
    pre_no_issue = max(0, 100 - pre_means.sum())
    post_no_issue = max(0, 100 - post_means.sum())
    
    # Add "No Issue" to the data
    pre_means = list(pre_means.values) + [pre_no_issue]
    post_means = list(post_means.values) + [post_no_issue]
    all_categories = categories + ['No Issue']
    
    # Print the category percentages for Pre-GenAI and Post-GenAI
    print("Pre-GenAI Distribution:")
    for cat, pct in zip(all_categories, pre_means):
        print(f"- {cat}: {pct:.1f}%")
    
    print("\nPost-GenAI Distribution:")
    for cat, pct in zip(all_categories, post_means):
        print(f"- {cat}: {pct:.1f}%")
    
    # Professional color palette (Tol muted scheme with 12 colors)
    colors = [
        '#EF5350',  # Red
        '#EC407A',  # Pink
        '#AB47BC',  # Purple
        '#5C6BC0',  # Indigo
        '#29B6F6',  # Light Blue
        '#26A69A',  # Teal
        '#66BB6A',  # Green
        '#D4E157',  # Lime
        '#FFEE58',  # Yellow
        '#FFA726',  # Orange
        '#FF7043',  # Deep Orange
        '#8D6E63'   # Brown
    ]
    
    # Rest of the function remains the same...
    # Custom autopct function with better formatting
    def custom_autopct(pct):
        return f"{pct:.1f}%" if pct > 1 else ""
    
    # Create figure with constrained layout
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8), 
                                  constrained_layout=True)
    
    # Pre pie chart
    wedges1, texts1, autotexts1 = ax1.pie(
        pre_means,
        labels=None,
        autopct=custom_autopct,
        startangle=90,
        pctdistance=0.8,
        colors=colors[:len(pre_means)],
        wedgeprops={'linewidth': 0.5, 'edgecolor': 'white'},
        textprops={'fontsize': 9, 'fontweight': 'bold'}
    )
    ax1.set_title('Pre-GenAI Distribution', pad=20, 
                 fontsize=12, fontweight='bold')
    
    # Post pie chart
    wedges2, texts2, autotexts2 = ax2.pie(
        post_means,
        labels=None,
        autopct=custom_autopct,
        startangle=90,
        pctdistance=0.8,
        colors=colors[:len(post_means)],
        wedgeprops={'linewidth': 0.5, 'edgecolor': 'white'},
        textprops={'fontsize': 9, 'fontweight': 'bold'}
    )
    ax2.set_title('Post-GenAI Distribution', pad=20, 
                 fontsize=12, fontweight='bold')
    
    # Create a single legend
    fig.legend(
        wedges1,
        all_categories,
        title="Issue Categories",
        loc="center right",
        bbox_to_anchor=(1.1, 0.5),
        fontsize=10,
        frameon=False
    )
    
    # Main title
    fig.suptitle('Comparison of Issue Category Distributions', 
                y=1.05, fontsize=14, fontweight='bold')
    
    plt.savefig('charts/category_pie_comparison.pdf',
            bbox_inches='tight',
            facecolor='white')
    plt.show()


def main():
    # Load and sample data
    # pre_data, post_data = load_and_sample_data(
    #     'data/pre_javascript_repository_issues__.csv',
    #     'data/post_javascript_repository_issues__.csv',
    #     sample_size=160
    # )
    pre_data = pre_merged_df
    post_data = post_merged_df
    
    print(f"Sampled {len(pre_data)} pre projects and {len(post_data)} post projects")
    
    # Generate comparison plots
    plot_category_comparison(pre_data, post_data)
    # plot_top_categories(pre_data, post_data)
    plot_category_pie_charts(pre_data, post_data)

if __name__ == "__main__":

    main()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib import rcParams

# Set random seed for reproducibility
np.random.seed(42)

# Set style for better looking plots
sns.set_style("whitegrid")
rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

def load_and_sample_data(pre_file, post_file, sample_size=160):
    """Load both CSVs and sample the same number of projects from each"""
    pre_df = pd.read_csv(pre_file)
    post_df = pd.read_csv(post_file)
    
    # Sample the data
    pre_sample = pre_df.sample(min(sample_size, len(pre_df)), random_state=42)
    post_sample = post_df.sample(min(sample_size, len(post_df)), random_state=42)
    
    return pre_sample, post_sample

def get_sorted_categories(pre_data, post_data):
    """Get alphabetically sorted list of category names"""
    categories = [col.replace(' %', '') for col in pre_data.columns if '%' in col]
    return sorted(categories)

def plot_category_comparison(pre_data, post_data):
    """Compare category distributions between pre and post with consistent ordering"""
    # Get sorted categories
    categories = get_sorted_categories(pre_data, post_data)
    category_cols = [f"{cat} %" for cat in categories]
    
    pre_means = pre_data[category_cols].mean()
    post_means = post_data[category_cols].mean()
    
    # Create comparison dataframe
    comparison_df = pd.DataFrame({
        'Category': categories,
        'Pre': pre_means.values,
        'Post': post_means.values
    })
    
    # Melt for seaborn
    melted_df = pd.melt(comparison_df, id_vars='Category', var_name='Period', value_name='Percentage')
    
    # Plot with consistent category order
    plt.figure(figsize=(14, 8))
    ax = sns.barplot(
        x='Category', 
        y='Percentage', 
        hue='Period', 
        data=melted_df,
        order=categories  # Enforce consistent ordering
    )
    plt.title('Average Issue Category Distribution Comparison (Pre vs Post)', pad=20)
    plt.xticks(rotation=45, ha='right')
    plt.ylabel('Average Percentage (%)')
    plt.legend(title='Analysis Period')
    plt.tight_layout()
    
    # Add value labels (only for non-zero values)
    for p in ax.patches:
        height = p.get_height()
        if height > 0.1:  # Only label if > 0.1% to avoid 0.0% labels
            ax.annotate(f"{height:.1f}%", 
                       (p.get_x() + p.get_width() / 2., height), 
                       ha='center', va='center', 
                       xytext=(0, 7), 
                       textcoords='offset points',
                       fontsize=10)
    
    plt.savefig('charts/category_comparison.png', dpi=300, bbox_inches='tight')
    plt.show()

def plot_top_categories(pre_data, post_data):
    """Compare top categories between pre and post with consistent ordering"""
    # Count top categories occurrences
    def count_top_cats(df):
        all_cats = []
        for cats in df['Top Categories'].str.split(', '):
            all_cats.extend(cats)
        return pd.Series(all_cats).value_counts(normalize=True) * 100
    
    pre_top = count_top_cats(pre_data)
    post_top = count_top_cats(post_data)
    
    # Get all unique categories alphabetically sorted
    all_categories = sorted(set(pre_top.index).union(set(post_top.index)))
    
    # Create comparison dataframe with consistent ordering
    comparison_df = pd.DataFrame({
        'Category': all_categories,
        'Pre': pre_top.reindex(all_categories).fillna(0).values,
        'Post': post_top.reindex(all_categories).fillna(0).values
    })
    
    # Melt for seaborn
    melted_df = pd.melt(comparison_df, id_vars='Category', var_name='Period', value_name='Percentage')
    
    # Plot with consistent category order
    plt.figure(figsize=(14, 8))
    ax = sns.barplot(
        x='Category', 
        y='Percentage', 
        hue='Period', 
        data=melted_df,
        order=all_categories  # Enforce consistent ordering
    )
    plt.title('Top Categories Comparison (Pre vs Post)', pad=20)
    plt.xticks(rotation=45, ha='right')
    plt.ylabel('Percentage of Projects Where Category Was Top 3 (%)')
    plt.legend(title='Analysis Period')
    plt.tight_layout()
    
    # Add value labels (only for non-zero values)
    for p in ax.patches:
        height = p.get_height()
        if height > 0.1:  # Only label if > 0.1%
            ax.annotate(f"{height:.1f}%", 
                       (p.get_x() + p.get_width() / 2., height), 
                       ha='center', va='center', 
                       xytext=(0, 7), 
                       textcoords='offset points',
                       fontsize=10)
    
    plt.savefig('charts/top_categories_comparison.png', dpi=300, bbox_inches='tight')
    plt.show()

def plot_category_pie_charts(pre_data, post_data):
    """Create pie charts for category distributions with consistent ordering"""
    # Get sorted categories
    categories = get_sorted_categories(pre_data, post_data)
    category_cols = [f"{cat} %" for cat in categories]
    
    pre_means = pre_data[category_cols].mean()
    post_means = post_data[category_cols].mean()
    
    # Custom autopct function to hide 0% values
    def custom_autopct(pct):
        return f"{pct:.1f}%" if pct > 0.5 else ""
    
    # Create figure
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))
    
    # Pre pie chart - using sorted categories
    wedges1, texts1, autotexts1 = ax1.pie(
        pre_means,
        labels=None,
        autopct=custom_autopct,
        startangle=90,
        pctdistance=0.85,
        textprops={'fontsize': 10}
    )
    ax1.set_title('Pre-Analysis Issue Category Distribution', pad=20)
    
    # Post pie chart - using same sorted categories
    wedges2, texts2, autotexts2 = ax2.pie(
        post_means,
        labels=None,
        autopct=custom_autopct,
        startangle=90,
        pctdistance=0.85,
        textprops={'fontsize': 10}
    )
    ax2.set_title('Post-Analysis Issue Category Distribution', pad=20)
    
    # Create a single legend for both charts with sorted categories
    fig.legend(
        wedges1,
        categories,
        title="Categories",
        loc="center right",
        bbox_to_anchor=(1.1, 0.5),
        fontsize=10
    )
    
    plt.suptitle('Comparison of Issue Category Distributions', y=1.05, fontsize=14)
    plt.tight_layout()
    plt.savefig('charts/category_pie_comparison.png', dpi=300, bbox_inches='tight')
    plt.show()

def main():
    # Load and sample data
    # pre_data, post_data = load_and_sample_data(
    #     'data/pre_javascript_repository_issues__.csv',
    #     'data/post_javascript_repository_issues__.csv',
    #     sample_size=160
    # )
    pre_data = pre_merged_df
    post_data = post_merged_df
    
    print(f"Sampled {len(pre_data)} pre projects and {len(post_data)} post projects")
    
    # Generate comparison plots
    plot_category_comparison(pre_data, post_data)
    plot_top_categories(pre_data, post_data)
    plot_category_pie_charts(pre_data, post_data)

if __name__ == "__main__":
    main()
```

```python
from math import pi

categories = ['mean_days', 'median_days', 'p75_days', 'std_dev_days']
N = len(categories)

values_pre = pre_merged_df[categories].mean().values
values_post = post_merged_df[categories].mean().values

angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, np.append(values_pre, values_pre[0]), 'o-', label='Pre-era')
ax.plot(angles, np.append(values_post, values_post[0]), 'o-', label='Post-era')
ax.fill(angles, np.append(values_pre, values_pre[0]), alpha=0.1)
ax.fill(angles, np.append(values_post, values_post[0]), alpha=0.1)
ax.set_theta_offset(pi/2)
ax.set_theta_direction(-1)
ax.set_thetagrids(np.degrees(angles[:-1]), categories)
plt.legend(loc='upper right')
plt.title('Multi-Metric Comparison', pad=20)
plt.show()
```
